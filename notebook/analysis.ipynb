```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
from pathlib import Path
```

```python
now = Path.cwd()
root = now.parent
path = root / "data" / "processed" / "train_processed.csv"

df = pd.read_csv(path)
df.head()
```

```python
plt.figure(figsize=(10,8))

sns.countplot(x='Loan_Status',data=df)
plt.xticks([0,1],["Non accord",'accord'])
```

```python

ax = sns.countplot(x='Loan_Status',data=df,hue='Credit_History')
plt.xticks([0,1],["Non accord",'accord'])

ax.legend(title='Credit_History', labels=['Non', 'Oui'])
```

```python
ax = sns.countplot(x='Loan_Status',data=df,hue='Gender')
plt.xticks([0,1],["Non accord",'accord'])

ax.legend(title='Gender', labels=['Female', 'Male'])
```

```python
ax = sns.countplot(x='Loan_Status',data=df,hue='Married')
plt.xticks([0,1],["Non accord",'accord'])

ax.legend(title='Married', labels=['No', 'yes'])
```

```python
ax = sns.countplot(x='Loan_Status',data=df,hue='Education')
plt.xticks([0,1],["Non accord",'accord'])

ax.legend(title='Education', labels=['Graduate', 'Non Graduate'])
```

```python
plt.figure(figsize=(10,10), dpi = 200)

sns.heatmap(df.corr(), annot=True, fmt='.2f')
```
